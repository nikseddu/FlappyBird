This assignment aims to train a simple agent to control the behaviour of the bird in the game of Flappy Bird.

We've checked that, once the all packages are installed correctly, the notebook can be executed on both MacOS or Windows.


There are two discrete **actions** in this environment:
- none (0)
- flap (1)

**Reward** is awarded as:
* +0.1 - every frame it stays alive
* +1.0 - successfully passing a pipe
* -1.0 - dying

And the environment offers two **observation** types.

1. A vector based observation where the elements in the vector are:
* the last pipe's horizontal position
* the last top pipe's vertical position
* the last bottom pipe's vertical position
* the next pipe's horizontal position
* the next top pipe's vertical position
* the next bottom pipe's vertical position
* the next next pipe's horizontal position
* the next next top pipe's vertical position
* the next next bottom pipe's vertical position
* player's vertical position
* player's vertical velocity
* player's rotation

2. An RGB image based observation of a screenshot of the game (not including score) of size 288 X 512 pixels. 

## Presentation of best agent

## 1. Vector Based Observation

### 1.1 Create and explore the environment

Create the environment.

```python
env = gym.make('FlappyBird-v0')
```

Explore the action space.

```python
env.action_space
```

Explore the observation space.

```python
env.observation_space
```

Play an episode of the environment using random actions.

```python
obs, _ = env.reset()
frames = []
while True:
    action = env.action_space.sample()
    obs, reward, terminated, done, info = env.step(action)

    # Rendering the game:
    env.render()
    time.sleep(1 / 30)  # FPS

    # Checking if the player is still alive
    if terminated:
        break
```

### 1.2 Train and test DQN agent

In this section, we train and test the DQN agent in this environment.

#### Training

Create the training environment and limit the max number of steps in an episode so that the agent won't run forever.

```python
# create training environment
train_env = gym.make("FlappyBird-v0")
# limit the max number of steps in an episode so that the bird won't fly forever
train_env = gym.wrappers.TimeLimit(train_env, max_episode_steps=3000)
```

Create the agent. A vanilla neural network is power enough for the vector observation so we use MlpPolicy.

```python
log_dir = "./dqn_vector_agent_log"

parameters = {
    "learning_rate": 0.0001,
    "buffer_size": 50000,
    "learning_starts": 5000,
    "batch_size":128,
    "gamma": 0.99,
    "exploration_fraction": 0.9,
    "target_update_interval": 1000,
    "exploration_initial_eps": 0.1,
    "exploration_final_eps": 0.0001,
}

# create agent
agent = sb3.DQN("MlpPolicy", train_env, tensorboard_log=log_dir, **parameters)
```

Examine the agent network architectures.

```python
print(agent.policy)
```

For this simple flappy bird application, the default network architecture, so we don't need to customized the network.

Create the environment for evaluation.

```python
# create evaluation environment
eval_env = gym.make("FlappyBird-v0")
eval_env = gym.wrappers.TimeLimit(eval_env, max_episode_steps=3000)
# create evaluation callback
eval_callback = sb3.common.callbacks.EvalCallback(
    eval_env,
    n_eval_episodes=10,
    eval_freq=5000,
    log_path=log_dir,
    best_model_save_path=log_dir,
    render=False,
)
```

Connect to the tensorboard log using **TensorBoard** from the command line to view training progress: 

`tensorboard --logdir ./dqn_vector_agent_log/`

Then open TensorBoard in a browser, typically located at:

`http://localhost:6006/`

We're only training for 10K steps  in this notebook to show the code works correctly. Actually, we trained the agent for 5 million steps on a linux server. Please refer to the train_scripts folder for full training on 5 million timesteps.

```python
agent.learn(total_timesteps=10000, callback=eval_callback)
```

The following cell plots the cumulative returns during the training process of 5 million steps. You can find the raw data in saved folder. 

```python
eval_log_path = 'saved/dqn_vector_agent_log/'
evaluation_log = np.load(eval_log_path + 'evaluations.npz')
evaluation_log_df = pd.DataFrame({item: [np.mean(ep) for ep in evaluation_log[item]] for item in evaluation_log.files})
ax = evaluation_log_df.loc[0:len(evaluation_log_df), 'results'].plot(color = 'lightgray', xlim = [-5, len(evaluation_log_df)], figsize = (10,5))
evaluation_log_df['results'].rolling(5).mean().plot(color = 'black', xlim = [-5, len(evaluation_log_df)])
ax.set_xticklabels(evaluation_log_df['timesteps'])
ax.set_xlabel("Eval Episode")
plt.ylabel("Rolling Mean Cumulative Return")
plt.show()
```

As the training progresses, the cumulative return increases, demonstrating a continuous improvement in the agent's performance. However, as we will see later, the results are not as good as others. This is due to the fact that, compared with image stack observation, vector observation cannot provide information for the agent to make decisioins.

### compare vector observation with image stack observation

By comparing two different types of observations, we have drawn several conclusions:

1. The image stack based agent outperformed the vector based agent, as expected. This is due to the richer information captured in the image stacks, which provide a more detailed representation of the game state.

2. The vector based agent, while not as effective as the image stack based agent, still demonstrated a reasonable ability to navigate the game. This indicates that the essential game mechanics can be learned and approximated from simple vector representations, but with lower overall performance.

3. The image stack observation may require more computational resources than the vector based observation due to the larger input size and complexity of the images. This trade-off should be considered when designing an agent for environments where computational resources are limited.

### compare DQN agent with PPO agent

By comparing two different types of agents, we have drawn several conclusions:

1. The DQN agent performed better than the PPO agent in our experiments. This result could be attributed to the inherent differences in the learning algorithms of the two agents. DQN, being an off-policy algorithm, uses a separate target network to estimate the Q-values, which can help stabilize learning and reduce overestimations of action values.

2. The PPO agent, while not as effective as the DQN agent, still managed to achieve reasonable performance. As an on-policy algorithm, PPO aims to strike a balance between exploration and exploitation by using a trust region optimization method. However, in the Flappy Bird environment, it appears that the DQN's learning algorithm was more suitable for the task.

3. It is possible that the performance difference between the DQN and PPO agents might be due to the choice of hyperparameters and network architectures. Further experiments with different configurations could potentially reveal additional insights and enhance the performance of both algorithms.

4. The DQN agent's success could also be related to the nature of the Flappy Bird environment, which is a discrete action space problem. DQN is generally known to be better suited for discrete action spaces, while PPO tends to work better in continuous action spaces.
